```python
# lets try performing merge and join ops, to do this lets create small dat frames
data1=pd.DataFrame({'key1':[1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]})
```

```python
data2=pd.DataFrame({'key1':[1,2,45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]})
```

```python
data1
```

```python
data2
```

```python
#now to perform merge
pd.merge(data1,data2)
#the way it works is that we have 2 dataframes, when we do merge operation, it compares values in columns of both dataframes 
# and whatever values which are common it merges
```

```python
#to perform left merge
pd.merge(data1,data2,how='left')
```

```python
#to perform right merge
pd.merge(data1,data2,how='right')
```

```python
#to perform outer merge
pd.merge(data1,data2,how='outer')
```

```python
#to perform cross merge
pd.merge(data1,data2,how='cross')
```

```python
#to perform left merge on a particular column(remember we cannot perform cross merge on columns, hence 
# doing on left,righ)
pd.merge(data1,data2,how='left', left_on = 'key2', right_on='key4')
```

```python
data1=pd.DataFrame({'key1':[1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]},
                  index=['a','b','c','d','e'])
```

```python
#lets do join operation,by default it will perform left join
data1.join(data2)
```

```python
#imp thing to note down is that incase of join it performs this operation based on indexes, and incase
#of merge it does based on column values 
```

```python
#lets do right join
data1.join(data2,how='right')
```

```python
#lets do outer
data1.join(data2,how='outer')
```

```python
#consider the above data set, here we have fare column, now we want to add a new column such that in that
#new column we have to convert the fare values into INR currency,right now its in euro
df["Fare_Inr"]=df["Fare"].apply(lambda x:x*80)
#this is achievable thru apply function
```

```python
df.head()
```

```python
#to find out length of name in the df
df['len_name']=df['Name'].apply(len)
```

```python
df.head()
```

```python
#other way of converting a currency like we did above is by writing a function
def in_conv(x):
    return x*80
df["Fare_Inr"]=df["Fare"].apply(in_conv)
```

```python
#another task we want to do is, we want to segregate the fares into maybe 3 type of categories, lets try creating a flag
def create_flag(x):
    if x<10:
        return "cheap"
    elif x>=10 and x<20:
        return "mid"
    else:
        return "high"
```

```python
df['Flag_fare']=df['Fare'].apply(create_flag)
```
