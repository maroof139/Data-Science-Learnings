```python
#lets import plotly library, its quite similar to matplotlib
import plotly.graph_objects as go
```

```python
#if we want to draw a scatter plot, we first create a figure(F is capital,), and then we add go.trace, and then write go.Scatter it is also
#important to mention the mode parameter, there are 2 types of modes-markers and line(like in what mode we want to plot our graph)
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5,3],y=[3,4,5,5,6,7],mode='markers'))
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5,3],y=[3,4,5,5,6,7],mode='lines'))
```

```python
#lets import seaborn library and load tips dataset
import seaborn as sns
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
#now with the help of plotly, lets plot a scatter plot with total_bill on x-axis and tips on y axis
#in order to plot a scatterplot, we first create a figure like this fig=go.Figure()
#then fig.add_trace(go.Scatter(x=any columns from dataset or whatever values to want to plot,y=any columns from dataset or whatever values to want to plot,
#mode=can be either 'markers' or 'lines'))
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
#lets try to change the mode to lines here just for example
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='lines'))
```

```python
#lets try to plot bar plot now, steps are same but here we dont need to pass the 'mode' marker param
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5,3],y=[3,4,5,5,6,7]))
```

```python
#lets try to plot a hostogram,we know in histogram, on x-axis it represents the data and on y-axis it represents the frequency
fig=go.Figure()
fig.add_trace(go.Histogram(x=[1,2,3,4,5,3]))
```

```python
#similarly lets plot histogram with respect to tips data for total_bill column
fig=go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))
#from the below histogram we can understand that there 1 person whose total bill is b/w 2 to 3.99 and so on..
```

```python
#here we have a scatter plot already representing tips.total_bill and tips.tips, so in this if we want to change the size of dots, so in this case we have a
#tips dataset and in that there is size columns, so if we also want to represent the size columns and based on that we can visualize various sizes with
#different sizes of dots we can use marker_style parameter and pass our column(this is how we do it - marker_size=tips['size])
#another way ofdoing the same thing is - marker_size=10*tips['size']
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size']))
#we can notice it is showing various size in diferent sizes of the dots
```

```python
#now if want to plot a 3d scatterplot for the same comparisons that is total_bill vs tips vs size, we can do this way
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',z=tips['size']))
#please notice that here instead of marker_size paraments we have changed it to z to represent size, because in 3d plot its imporatnt to have x,y and z values
```

```python
#now if want to plot a 3d scatterplot for the same comparisons that is total_bill vs tips vs size, we can do this way
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,z=tips['size'],mode='lines'))
#please notice that here instead of marker_size paraments we have changed it to z to represent size, because in 3d plot its imporatnt to have x,y and z values
```

```python
#lets try plotting a 3d plot again but this time with different dataset
#now if want to plot a 3d scatterplot for the same comparisons that is total_bill vs tips vs size, we can do this way
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4,2],y=[3,4,5,6,6],z=[3,4,5,6,7],mode='markers'))
#please notice that here instead of marker_size paraments we have changed it to z to represent size, because in 3d plot its imporatnt to have x,y and z values
```
