## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

### Matplotlib is a popular 2D plotting library for Python. It provides a wide variety of high-quality and customizable static, animated, and interactive plots for data visualization. Matplotlib is widely used in scientific computing, data analysis, machine learning, and other fields where visualizing data is essential.
### The main module within Matplotlib for creating static, animated, and interactive plots is called pyplot. Pyplot provides a simple and consistent interface for creating a wide range of plots. Some common types of plots that can be created using the Pyplot module of Matplotlib include:
### Line Plot: A basic plot that displays data points connected by straight line segments.
### Scatter Plot: Displays individual data points without connecting them, useful for visualizing the distribution of data.
### Bar Plot: Represents data with rectangular bars, where the length of each bar is proportional to the value it represents.
### Histogram: Represents the distribution of a continuous dataset. It divides the data into bins and shows the frequency of data points in each bin.
### Pie Chart: Represents data in a circular graph, where each wedge represents a proportion of the whole.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
### import numpy as np
### np.random.seed(3)
### x = 3 + np.random.normal(0, 2, 50)
### y = 3 + np.random.normal(0, 2, len(x))
### Note: Also add title, xlabel, and ylabel to the plot.

### A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane. In a scatter plot, each data point is represented by a dot, and the position of the dot on the graph is determined by its values along two numeric variables, usually one plotted along the x-axis and the other along the y-axis.
### Key characteristics of a scatter plot:
### Individual Data Points: Each data point is represented by a single dot on the plot.
### Two Variables: The plot visualizes the relationship between two variables. One variable is plotted along the horizontal axis (x-axis), and the other variable is plotted along the vertical axis (y-axis).
### Patterns and Trends: Scatter plots are useful for identifying patterns, trends, correlations, or clusters within the data. They help to understand the relationship between the two variables.
### No Line Connectors: Unlike line plots, scatter plots do not connect the data points with lines. Each point stands alone, making it suitable for displaying the distribution of data.
### Data Point Markers: Each data point is often marked with a marker or symbol (such as a dot or circle), making it easy to distinguish individual points.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.title("This is the scatter plot for x and y")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:
### import numpy as np
### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

### The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid, specifying the number of rows and columns, and then select a specific subplot to plot in. This is useful when you want to compare different datasets or visualize multiple aspects of your data in a single figure.

```python
x=np.array([0, 1, 2, 3, 4, 5])
y1=np.array([0, 100, 200, 300, 400, 500])
y2=np.array([50, 20, 40, 20, 60, 70])
y3=np.array([10, 20, 30, 40, 50, 60])
y4=np.array([200, 350, 250, 550, 450, 150])
#here plt.subplot(2,2,1) means 2 rows 2 columns and plot in position 1 in the grid 
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Plot 4")

#tight_layout function is used for better spacing between the grids
plt.tight_layout()

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
### import numpy as np
### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
### profit = np.array([3000, 8000, 1000, 10000])

### Answer: A bar plot is a type of data visualization that represents categorical data with rectangular bars. The length of each bar is proportional to the values it represents. Bar plots are commonly used for visualizing the distribution of data across different categories or groups. The x-axis typically represents the categories, and the y-axis represents the values or frequencies associated with each category.

### Key features and uses of bar plots:

### Categorical Data: Bar plots are especially useful when dealing with categorical data, where each bar corresponds to a distinct category or group.
### Comparisons: Bar plots make it easy to compare the values of different categories, as the lengths of the bars are directly comparable. This is helpful for identifying trends, patterns, or differences in the data.
### Frequency Distribution: Bar plots can display the frequency distribution of categorical variables, showing how often each category occurs.
### Visualizing Relationships: Bar plots can be used to visualize relationships between different categorical variables. For example, you might use grouped or stacked bar plots to compare subcategories within each main category.
#### Nominal or Ordinal Data: Bar plots are suitable for both nominal and ordinal data, where the categories have no inherent order (nominal) or have a defined order (ordinal).

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,color='#6a5acd')
```

```python
#To plot a horizontal Bar Plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit,color='#6a5acd')
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
### box1 = np.random.normal(100, 10, 200)
### box2 = np.random.normal(90, 20, 200)

### Answer: A box plot, also known as a box-and-whisker plot, is a statistical visualization that displays the distribution and summary statistics of a dataset. It provides a concise and informative way to represent the central tendency, spread, and potential outliers in the data. The key features of a box plot include:
### Box (Interquartile Range): The box in the plot represents the interquartile range (IQR), which is the middle 50% of the data. The bottom and top edges of the box represent the first quartile (Q1) and the third quartile (Q3), respectively. The length of the box is the IQR.
### Whiskers: The whiskers extend from the edges of the box to indicate the range of the data. They are often defined to a certain multiple of the IQR, and points beyond the whiskers may be considered as potential outliers.
### Median Line: A line inside the box represents the median (Q2), which is the middle value of the dataset when it is sorted.
### Outliers: Individual data points beyond the whiskers are considered outliers and are often plotted individually.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],labels=['Box 1','Box 2'])
plt.xlabel("Boxes")
plt.ylabel("Value")
plt.title("This is Box Plot")
```
