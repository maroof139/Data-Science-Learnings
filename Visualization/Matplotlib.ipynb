```python
#lets import matplotlib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#now lets genrate random number using  np.random.rand
x=np.random.rand(50)
#here 50 means we want to generate 50 datapoints5
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
#to create a scatter plot we are using the above x and y data points, to make scatter plot, this is how we do
plt.scatter(x,y,c='r',alpha=0.5)#if we want to control the color, we simply mention it just after plt.scatter(x,y,c='r'),
#c means color as shown above 
#if we want to increase or decrease the intensity of the color, we use the keyword called alpha
#if we want to write something in x and y axis, we use plt.xlabel and plt.ylabel
plt.xlabel("this is x axis")
plt.ylabel("this is y label")
#if we want to give title for the graph
plt.title("this is x vs y")
```

```python
#if we want to control size of the graph or grid, we use keyword called plt.figure(figsize=())
plt.figure(figsize=(6,2))
#to create a scatter plot we are using the above x and y data points, to make scatter plot, this is how we do
plt.scatter(x,y,c='r',alpha=0.9,marker='x')#if we want to control the color, we simply mention it just after plt.scatter(x,y,c='r'),
#c means color as shown above 
#if we want to increase or decrease the intensity of the color, we use the keyword called alpha, and it should be 
#in range between 0-1 as shown above 
#if we want to change the marker, that is instead of circular dots, i want to display it as x, we use keyword called marker

#if we want to write something in x and y axis, we use plt.xlabel and plt.ylabel
plt.xlabel("this is x axis")
plt.ylabel("this is y label")
#if we want to give title for the graph
plt.title("this is x vs y")

#if we want to display the above graph in the form of grid
plt.grid()
```

```python
#6,3lets control the figure size(this code should be in top or else it will throw error)
plt.figure(figsize=(5,3))
#lets generate a bar plot using plt.bar
#to add a different color in bar plot, name is color instead of just 'c'
plt.bar(x,y,color='r')
#lets set x label and y label and add title
plt.xlabel("this is categorical data")
plt.ylabel("this is numerical data")
plt.title("this is bar plot")
#lets put this in a grid
plt.grid()
```

```python
#6,3lets control the figure size(this code should be in top or else it will throw error)
plt.figure(figsize=(5,3))
#lets generate a bar plot using plt.bar
#to add a different color in bar plot, name is color instead of just 'c'
#the above graph was in vertical, if we want the graph to be in horizontal way, we do it by using plt.barh
plt.barh(x,y,color='r')
#x axis has got rotated and became y axis, and y axis has rotated and became x axis 
#lets set x label and y label and add title
plt.xlabel("this is numerical data")
plt.ylabel("this is categorical data")
plt.title("this is bar plot")
#lets put this in a grid
plt.grid()
```

```python
data=[1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
#As we can see there is repetition in 'data' above. lets create a plot such that on x axis it should plot the 
#'data' and on y axis it should plot the frequency, meaning in x axis it should plot the unique data in 'data' and on 
#y axis it should plot how many times the repetition has happened or how many times the number has got repeated
#this type of plotting is possible with the help of histogram plot using plt.hist(data)
plt.figure(figsize=(6,3))
plt.hist(data,color='#6a5acd')
#we can see that on x axis, we can see all the distinct data and on y axis it is showing frequency of the data(meaning how 
#many times its got repeated)
#let understand the graph, 1 2 3 has appeared only 1 time hence it has plotted ((1,1),(2,1),(3,1)),(1,2,3 are on x axis and 
#1 1 1 is on y axis) 4 has appeared once, 5 has appeared 2 times hence(5,2), 6 appeared 3 times hence (6,3),7 appeared 2 times
#hence (7,2), 8 appeared 5 times hence (8,5)
#lets add x and y label and title
plt.xlabel("this represents unique data")
plt.ylabel("this represents frequency of data")
plt.title("Histogram plot example")
#lets put it in a grid
plt.grid()
#if we want to save the plot as a png file
plt.savefig("test.png")
```

```python
#lets create a 3d plot
#for that lets generate data using np.random.rand(no of data points we want to generate)
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
#for creating a 3d plot first we need to create a figure and store it in a variable
fig=plt.figure()
#now inside the we need to define, meaning we need to add subplot inside the fig(fig variable contains our figure)
#when we are adding subplot inside the figure,its important to mention projection(meaning type of projection), in this case 
#it will be 3d
#we are placing the subplot inside a variable(variable name could be anything)
ax=fig.add_subplot(projection='3d')
#in case of 3d plots, we can add color using 'c' and also we can control intensity of dots using alpha keyword
ax.scatter(x,y,z,c='#6a5acd',alpha=.9)
#lets add labels
plt.xlabel("this is x-axis")
plt.ylabel("this is y-axis")
#its important to call plt.show(), without this plot will not be displayed
plt.show()
#to generate a 3d plot its important to add these 2 steps:-
#fig=plt.figure()
#ax=fig.add_subplot(projection='3d')
```
