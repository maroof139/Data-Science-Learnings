## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Answer. Seaborn is a powerful data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive statistical graphics. Here are five common plots that you can create using Seaborn along with their uses:

### Scatter Plot:
### Use: Display the relationship between two numerical variables.
### Seaborn Function: sns.scatterplot(x, y, data)

### Line Plot:
### Use: Show the trend of a numerical variable over a continuous interval or time.
### Seaborn Function: sns.lineplot(x, y, data)

### Bar Plot:
### Use: Display the distribution of a categorical variable and compare values across different categories.
### Seaborn Function: sns.barplot(x, y, data)

### Histogram:
### Use: Visualize the distribution of a single numerical variable.
### Seaborn Function: sns.histplot(data, x, bins)

### Heatmap:
### Use: Display the correlation between multiple variables in a matrix form.
### Seaborn Function: sns.heatmap(data, annot, cmap)
### These are just a few examples, and Seaborn provides many more plotting functions for various types of visualizations. The choice of the plot depends on the nature of your data and the insights you want to extract.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
fmri['region'].value_counts()
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri,hue='region',style='event')
```

```python
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='plasma', annot=True,fmt='d')

# Set plot title and labels
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
